Here is a simple example of gradient descent in one dimension using Python:

Suppose we want to find the minimum of a convex function f(x) = x^2. We can use gradient descent to find the minimum by iteratively updating the value of x as follows:


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to minimize
def f(x):
    return x**2

# Define the derivative of the function
def df(x):
    return 2*x

# Set the initial value of x
x = 5

# Set the learning rate
alpha = 0.1

# Set the number of iterations
num_iterations = 100

# Perform gradient descent
x_values = [x]
for i in range(num_iterations):
    # Compute the gradient
    gradient = df(x)
    
    # Update x
    x = x - alpha * gradient
    
    # Add x to the list of x values
    x_values.append(x)

# Plot the function and the path taken by gradient descent
x_range = np.linspace(-6, 6, 100)
y_range = f(x_range)
plt.plot(x_range, y_range, label="f(x)")
plt.plot(x_values, [f(x) for x in x_values], 'ro-', label="gradient descent")
plt.legend()
plt.show()
```

In above example, we first import the NumPy and Matplotlib libraries to help us with the plotting. We define the function f(x) = x^2 and its derivative df(x) = 2x, and set the initial value of x to 5, the learning rate alpha to 0.1, and the number of iterations to 100. We then perform gradient descent by computing the gradient df(x) at each iteration, updating x using the formula x = x - alpha * df(x), and adding the current value of x to a list of x values.

Finally, we plot the function f(x) and the path taken by gradient descent using the Matplotlib library. The red dots represent the values of x visited by gradient descent, and we can see that the algorithm successfully finds the minimum of the function at x = 0.
